4. **Structure and Operation**

 *(Revenue ‚Üí Expense ‚Üí Stock Purchase ‚Üí Products ‚Üí Sales ‚Üí Revenue)*

 4.1 Initial Capital

 The system starts with a value of $100.000.00 entered in the Revenue table, representing the capital available in the market's cash register.

 4.2 Expense Generation

 Expenses are automatically generated based on the available capital and recorded in the Expense table.

 4.3 Inventory Purchases

 These expenses feed the Inventory Purchase table, which records the products purchased (type, price, quantity). They are entered in the Available Products table, with the final price already calculated based on the defined profit margin.

 4.4 Sales Simulation
 Through Python automation, the system generates simulated sales (invoices). These transactions are analyzed and the data is returned to the Revenue table, closing the cycle.

 4.5 Operational Cycle
 The cycle (Revenue ‚Üí Expense ‚Üí Stock Purchase ‚Üí Products ‚Üí Sales ‚Üí Revenue) is continuous and can be executed as many times as necessary, simulating the financial and operational flow of a real market.
